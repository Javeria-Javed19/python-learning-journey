Question:

Write a Python program that uses a while True loop with if-elif-else conditions to modify a string using slicing and update a counter. The loop should stop using break and then print the final string and count.

Simple Description of the Code

This Python code demonstrates the use of a while loop, if-elif-else conditions, string indexing, and string slicing.

At the start, a string "aabbcc" and a variable count with value 3 are created.
The while True loop means the loop will run continuously until a break statement stops it.

Inside the loop:
Inside the loop:

The if condition checks the first character of the string.
If it is 'a', the first two characters are removed using string slicing.

The elif condition checks the last character of the string.
If it is 'b', only the first two characters of the string are kept.

If both conditions are false, the else block runs.
In this block, the counter is increased by 1 and the break statement stops the loop.

Finally, the program prints the updated string and the final value of count.

This code helps in understanding:

Infinite loops (while True)

Conditional statements (if-elif-else)

String indexing and slicing

Loop control using break.

```python
string = "aabbcc"
count = 3

while True:
    if string[0] == 'a':
        string = string[2:]
    elif string[-1] == 'b':
        string = string[:2]
    else:
        count += 1
        break

print(string)
print(count)
```

Question:

Write a Python program that uses a while True loop with if-elif-else conditions to modify a string using slicing and update a number variable. The loop should stop using break and then print the final string and number.

Simple Description of the Code

This Python program shows the use of a while loop, conditional statements, and string slicing.

At the beginning, a string "ccddaa" and a variable num with value 2 are created.
The while True loop means the loop will keep running again and again until a break statement stops it.

Inside the loop:

The if condition checks the first character of the string.
If it is 'c', one character is removed from the start of the string.

The elif condition checks the last character of the string.
If it is 'a', one character is removed from the end of the string.

If both conditions are false, the else block runs.
In this block, the number variable is increased and the loop stops using break.

At the end, the program prints the updated string and the final value of the number.



```python
text = "ccddaa"
num = 2

while True:
    if text[0] == 'c':
        text = text[1:]
    elif text[-1] == 'a':
        text = text[:-1]
    else:
        num += 2
        break

print(text)
print(num)
```

QUESTION

WHAT ARE BREAK AND CONTINUE STATEMENTS.

Break Statement

The break statement is used inside loops to immediately stop the execution of the loop.
When the break statement is reached, the loop ends and the program continues with the next line after the loop.
It is useful when a condition is met and there is no need to continue the loop further.

Continue Statement

The continue statement is used to skip the current iteration of a loop and move to the next iteration.
It does not stop the loop completely; it only ignores the remaining code in the current cycle.
It is helpful when we want to avoid certain values but still continue the loop.
