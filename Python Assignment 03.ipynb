MATCH CASE STATEMENTS

The match case statement is used when we want to compare one value with many fixed options.
It helps us to avoid writing very long if-elif chains and makes code easier to read.

IMPORTANT POINTS:

. match checks a variable.

. case represnts each possible value.

. _underscore work like a default option.

. Mostly used in menus,categories,options.


General Syntax

match variable_name:

    case value1:
        #action
    case value2:
        #action
    case _:
        # default action
        

PROGRAMS USING MATCH CASE

EXAMPLE 01: WEATHER ADVISOR

DESCRIPTION:

This program suggests an activity based on the weather.

```python
weather= input("ENTER WEATHER(sunny/rainy/cold):") . lower()
match weather:
    case "sunny" :
        print("You can go for a walk.")
    case "rainy":
        print("Take an umbrella.")
    case "cold":
        print("Wear warm clothes.")
    case _:
        print("Weather not recognized.")
        
```

EXAMPLE 02: Pet Sound Program


DESCRIPTION:

User enters an animal and program print its sound.

```python
animal = input("ENTER ANIMAL(cat/dog/cow):").lower()
match animal:
  case "cat":
    print("Meow Meow")
  case "dog":
    print("Bark Bark")
  case "cow":
    print("MOO MOO")
  case _:
    print("Unknown Animal")
    
```

EXAMPLE 03: Transport Selector

DESCRIPTION:

User choose a tranport mode and program responds.

```python
transport = input("Choose transport (bus/train/bike):").lower()
match transport:
    case "bus":
        print("Public Transport Selected")
    case "train":
        print("Fast Travel Selected")
    case "bike":
        print("Eco Friendly Ride")
    case _:
        print("Invalid Option")
        
```

Difference between If-Else and Match case

If-Else Statements

. Used for logical conditions like >,<, ranges,and comparsions.

. Good for complex decisions making.

. Code can become long if there are many conditions.

. Conditions are checked one by one.

. Uses else as the default block.



Match Statement

. Used to compare one value with fixed options(exact wors or numbers).

. Best for menus, categories and shorter.

. Focuses on pattern/value matching instead of logic operators.

. Uses case _as the default options.



NESTED If-Else Statement

A Nested If-Else means placing one if inside another if.

It is useful when second decision depends on first decision.

EXAMPLE 01: MOVIE TICKET CHECK

```python
age = int(input("Enter age: "))

if age >= 13:
    student = input("Are you a student? (yes/no): ")
    if student == "yes":
        print("Student Discount Applied")
    else:
        print("Regular Ticket")
else:
    print("Too young for this movie")
```

Example 02: GYM MEMBERSHIP

```python
height = int(input("Enter height in cm: "))

if height >= 150:
    weight = int(input("Enter weight: "))
    if weight >= 45:
        print("You can join gym")
    else:
        print("Weight too low")
else:
    print("Height requirement not met")
    
```

EXAMPLE 03: INTERNET PACKAGE

```python
data_used = int(input("Enter data used (GB): "))

if data_used > 5:
    premium = input("Premium user? (yes/no): ")
    if premium == "yes":
        print("Unlimited Access")
    else:
        print("Extra Charges Applied")
else:
    print("Within Limit")
```

EXAMPLE 04:SCHOOL TRIP PERMISSION

```python
fee_paid = input("Fee paid? (yes/no): ")

if fee_paid == "yes":
    permission = input("Parent permission? (yes/no): ")
    if permission == "yes":
        print("Trip Allowed")
    else:
        print("Permission Missing")
else:
    print("Fee Required")
```

EXAMPLE 04: MOBILE UNLOCK SYSTEM

```python
pin = input("Enter PIN: ")

if pin == "0000":
    fingerprint = input("Fingerprint verified? (yes/no): ")
    if fingerprint == "yes":
        print("Phone Unlocked")
    else:
        print("Fingerprint Failed")
else:
    print("Wrong PIN")
```
